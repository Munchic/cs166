```python
import random
import matplotlib.pyplot as plt
```

```python
NUM_SIM = 10000
INIT_MONEY = 250
NUM_BETS = 20

head = 1
tail = 1 - head
p_head = 0.6
bet_frac = 0.2

outcome_store = []

for _ in range(NUM_SIM):
    cur_money = INIT_MONEY
    for i in range(NUM_BETS):
        bet = bet_frac * cur_money
        if random.random() < p_head:
            cur_flip = head
        else:
            cur_flip = tail

        if cur_flip == head:
            cur_money += bet
        else:
            cur_money -= bet
            
    outcome_store.append(cur_money)
```

```python
plt.figure(figsize=(16, 5), dpi=330)
plt.hist(outcome_store, bins=100, color='skyblue', density=True)
plt.xlabel('amount after 20 games')
plt.ylabel('probability density')
plt.axvline(INIT_MONEY, linestyle='--', color='red', label='initial sum=${0}'.format(INIT_MONEY))

plt.legend()
plt.show()
```

```python
CONF_INT = 0.95
outcomes_sorted = sorted(outcome_store)
confint_left = outcomes_sorted[int(NUM_SIM * (1 - CONF_INT) / 2)]
confint_right = outcomes_sorted[int(NUM_SIM * (1 + CONF_INT) / 2)]
print('95% confindence interval: [{0}, {1}]'.format(confint_left, confint_right))
```
