```python
import numpy as np
import pandas as pd

import cvxpy as cp
```

```python
berlin_time = np.array([
    [0, 54, 27, 41, 33, 39, 85, 63],
    [63, 0, 26, 15, 20, 33, 43, 22],
    [32, 24, 0, 13, 5, 16, 59, 36],
    [46, 16, 11, 0, 6, 22, 48, 26],
    [40, 19, 4, 8, 0, 16, 52, 30],
    [41, 32, 15, 26, 18, 0, 70, 48],
    [103, 52, 67, 58, 63, 79, 0, 26],
    [70, 20, 35, 25, 30, 47, 24, 0],
])

np.diag(berlin_time)
```

```python
n = 8

x = cp.Variable((n, n), integer=True)

cts = []

objective_function = cp.sum(cp.multiply(berlin_time, x))
obj = lambda x: np.sum(berlin_time * x)
constraints = [
    cp.sum(x, axis=0) == np.ones(n),
    cp.sum(x, axis=1) == np.ones(n),
    x >= 0,
    x[4][3] + x[5][3] + x[0][3] + x[2][3] + x[4][1] + x[5][1] + x[0][1] + x[2][1] >= 1,
    x[4][6] + x[5][6] + x[0][6] + x[2][6] + x[4][7] + x[5][7] + x[0][7] + x[2][7] >= 1,
    
    # *list(x[i][j] * x[j][i] != 1 for i, j in (list(combinations(range(n), 2)) + list(combinations(list(range(n))[::-1], 2)))),
    cp.diag(x) == 0
]

prob = cp.Problem(cp.Minimize(objective_function), constraints)
prob.solve(solver='ECOS_BB')

# Print result.
print(f"The LP optimal value (and upper bound) is {prob.value:.4f}")
print("The LP solution of x is")
print(np.abs(np.round(x.value)))
```

```python
path = np.abs(np.round(x.value))
import networkx as nx
import matplotlib.pyplot as plt

nx_graph = nx.convert_matrix.from_numpy_matrix(np.abs(np.round(x.value)), parallel_edges=True, create_using=nx.MultiDiGraph())
pos = nx.circular_layout(nx_graph)
nx.draw_circular(nx_graph)
labels = {i : i for i in nx_graph.nodes()}
nx.draw_networkx_labels(nx_graph, pos, labels, font_size=15, font_color = 'white')

G = nx.from_numpy_matrix(np.matrix(berlin_time * path), create_using=nx.DiGraph)
nx.draw_networkx_edge_labels(G, pos=pos)

plt.show()
```
