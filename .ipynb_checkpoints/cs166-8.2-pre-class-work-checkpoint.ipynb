```python
import networkx as nx
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
g = nx.erdos_renyi_graph(30, 0.05, directed=True, seed=123)
nx.draw(g, pos=nx.kamada_kawai_layout(g))
```

```python
adj_mat = np.array(nx.adjacency_matrix(g).todense(), dtype='float32')

for j in range(adj_mat.shape[1]):
    col_sum = np.sum(adj_mat[:, j])
    if np.sum(col_sum) > 0:
        adj_mat[:, j] = adj_mat[:, j] / col_sum
    else:
        adj_mat[:, j] = 1 / adj_mat.shape[0]
```

```python
def rand_surf_sim(N=1000, alpha=0.85):
    all_nodes = list(g.nodes())
    cur_node = random.choice(all_nodes)
    visits = np.zeros(adj_mat.shape[0])

    for i in range(N):
        visits[cur_node] += 1
        adj_nodes = list(g.neighbors(cur_node))

        if len(adj_nodes) > 0 and random.random() < alpha:
            cur_node = random.choice(adj_nodes)
        else:
            cur_node = random.choice(all_nodes)

    visits /= np.sum(visits)
    
    return visits
```

```python
def err(pagerank_visits, sim_visits):
    return np.sqrt(np.sum((pagerank_visits - sim_visits)**2))

pgrk_visits = np.array([nx.pagerank(g)[key] for key in nx.pagerank(g)])
```

```python
n_range = np.arange(100, 10001, 100)
err_store = []

for n in n_range:
    visits = rand_surf_sim(N=n)
    err_store.append(err(pgrk_visits, visits))
```

```python
plt.figure(figsize=(16, 6), dpi=330)
plt.plot(n_range, err_store)
plt.ylabel('RMSE')
plt.xlabel('# iterations')
plt.show()
```
