```python
data = pd.read_csv('14.2-proctatinium-data.csv')
```

```python
plt.figure(figsize=(16, 6), dpi=330)
plt.plot(data['time'], data['count_rate'])
plt.ylabel('count rate')
plt.xlabel('time [s]')
plt.show()
```

```python
def decay_fn(_lambda, t, N_0=32):
    return N_0 * np.exp(-_lambda * t)
```

```python
NUM_TRIALS = 10000
lambda_store = []

for _ in range(NUM_TRIALS):
    count_noise = np.random.normal(size=data['count_rate'].size)
    new_count = data['count_rate'].to_numpy() + count_noise
    times = data['time'].to_numpy()
    
    init_lambda = 1
    lambda_store.append(curve_fit(decay_fn, times, new_count, init_lambda)[0][0])
```

```python
plt.figure(figsize=(16, 6), dpi=330)
plt.hist(lambda_store, bins=20)
plt.ylabel('probability density')
plt.xlabel('lambda')
plt.show()
```

```python
print('error = {0} s'.format(np.log(2) / np.mean(lambda_store) - 60 * 1.17))
```

This is not MCMC method because we are not trying to sample a random parameter and accept/reject it. Instead, we use curve-fitting which is just minimizing the least squares error.
